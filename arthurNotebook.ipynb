3.1.2 Propriedades da integrais triplas

se $F(x,y,z)$ é uma função definida em uma região D fechada e limitada no espaço, como a região ocupada por uma bola sólida ou uma pelota de argila, então a integral de $F$ sobre $D$ pode ser definida como:

$$\iiint\limits_DF(x,y,z)dxdydz$$

#### 3.1.3 Volume de uma região no espaço

**Definição**: o **volume** de uma região $D$ fechada e limitada no espaço é

$$\iiint\limits_DdV$$

onde $dV=dxdydz$

Exemplo
(3.1.3.)(a) $$\int_0^1\int_0^1\int_0^1(x² +y² +z²)dzdydx  $$

*Solução*

Dado a integral $$\int_0^1\left[\int_0^1\left[\int_0^1(x² +y² +z²)dz\right]dy\right]dx$$ calculamos primeiro a parte de dentro em relação a $z$, com $x$ e $y$ constantes temos

$$\int_0^1\left[\int_0^1\left[(x²z +y²z +\frac{z³}{3})\right]_0^1dy\right]dx$$ 
subistituindo o intervalo $[0,1]$ temos
$$\int_0^1\left[\int_0^1(x² +y² +\frac{1}{3})dy\right]dx$$ 

calucando agora em relação a $y$, com $x$ constante, temos

$$\int_0^1\int_0^1\left[(x²y +\frac{y³}{3} +\frac{1}{3}y)\right]_0^1dx$$
subistituindo o intervalo $[0,1]$ temos
$$\int_0^1(x² +\frac{1}{3} +\frac{1}{3})dx = \int_0^1(x² +\frac{2}{3})dx $$
por fim, calculamos em relação a $x$, então temos

$$\left[\frac{x³}{3} +\frac{2}{3}x\right]_0^1 = (\frac{1}{3} +\frac{2}{3}) = 1$$

portanto $$\int_0^1\int_0^1\int_0^1(x² +y² +z²)dzdydx = 1  $$

$$\int_{0}^{\sqrt2}\int_{0}^{3y}\int_{x²+3y²}^{8-x²-y²}dzdxdy =  \int_{0}^{\sqrt2}\int_{0}^{3y}\left[z\right]_{x²+3y²}^{8-x²-y²}dxdy  =\int_{0}^{\sqrt2}\int_{0}^{3y}(8-x²-y²) - (x²+3y²)dxdy = \int_{0}^{\sqrt2}\int_{0}^{3y}(8 - x²-y²- x²- 3y²)dxdy =\int_{0}^{\sqrt2}\int_{0}^{3y}(-2x²-4y²+8)dxdy  $$

Calculando em relação a $x$, com $y$ constante
$$\int_{0}^{\sqrt2}\int_{0}^{3y}(-2x²-4y²+ 8)dxdy =\int_{0}^{\sqrt2}\left[\frac{-2x³}{3}-4y²x+8x\right]_{0}^{3y}dy = \int_{0}^{\sqrt2}\left(\frac{-2(3y)³}{3}-4y²(3y)+8(3y)\right) - \left(\frac{-2(0)³}{3}-4y²(0)+8(0)\right)dy =\int_{0}^{\sqrt2}(-18y³-12y³+24y)dy = \int_{0}^{\sqrt2}(-30y³+24y)dy  $$

$$\int_{0}^{\sqrt2}(-30y³+24y)dy = [\frac{-15y⁴}{2} + 12y² ]_{0}^{\sqrt2} = (\frac{-15(\sqrt2)⁴}{2} + 12(\sqrt2)²) - (\frac{-15(0)⁴}{2} + 12(0)²) = -30  + 24 = -6 $$



# Como Calcular a integral  tripla no Python

Abaixo temos o algoritmo de como resolver uma integral tripla utilizando o $Python$:


```python
#Exemplo 3.1.3 (a)

from __future__ import division # Importando as Bibliotecas
from sympy import *
init_printing()  
x, y, z = symbols('x y z') #define x, y e z como variáveis simbólicas.

def f(x,y,z):return x**2 +y**2 + z**2 #definindo a funçao na qual vai integrar

#para integrar por partes
Emz= integrate(f(x,y,z), (z,0,1))#calculando a integral primeiro no intervalos 0<=z<=1
print('integral em 0<=z<=1\n\n',Emz)
Emy = integrate(Emz, (y,0,1)) #calculando a integral primeiro no intervalos 0<=y<=1 
                        #ultilizando a resposta Emz anterior
print('\nintegral em 0<=y<=1\n\n',Emy)
Emx = integrate(Emy, (x,0,1)) #calculando a integral primeiro no intervalos 0<=x<=1 
                            #ultilizando a resposta Emy anterior
print('\nintegral em 0<=x<=1\n\n',Emx)

#para integrar com todos os intervasos definidos

resposta = integrate(f(x,y,z), (z,0,1),(y,0,1),(x,0,1))

print('\nintegral em todos os intervalos\n\n',resposta)
```

```python
#Exemplo (3.1.3.)(b)  
from __future__ import division # Importando as Bibliotecas
from sympy import *
init_printing()

def f1(x,y,z):return 1 #definindo a funçao na qual vai integrar

#para integrar por partes
Emz= integrate(f1(x,y,z), (z,x**2 + 3*y**2,8 - x**2 - y**2))#calculando a integral primeiro no intervalos 0<=z<=1
print('integral em x² + 3y²<=z<=8 - x² - y²\n\n',Emz)


Emx = integrate(Emz, (x,0,3*y)) #calculando a integral primeiro no intervalos 0<=x<=3y 
                            #ultilizando a resposta Emz anterior
print('\nintegral em 0<=x<=3y\n\n',Emx)


Emy = integrate(Emx, (y,0,sqrt(2))) #calculando a integral primeiro no intervalos 0<=y<=raiz(2) 
                            #ultilizando a resposta Emx anterior
print('\nintegral em 0<=y<=raiz(2)\n\n',Emy)

#para integrar com todos os intervasos definidos

resposta = integrate(f1(x,y,z), (z,x**2 + 3*y**2,8 - x**2 - y**2),(x,0,3*y),(y,0,sqrt(2)))

print('\nintegral em todos os intervalos\n\n',resposta)
```

#### 3.1.4 Momentos e centros de massa

**Sólido Tridimensional**

**Massa:** $ M = \iiint\limits_D\delta dV \ \ \text{onde } \delta=\delta(x,y,z) \text{ é a densidade em }(x,y,z)$

**Primeiros momentos em relação aos planos coordenados**

$$M_{yz} = \iiint\limits_Dx\delta dV \text{, }\ \ M_{xz} = \iiint\limits_Dy\delta dV \text{,}\ \ M_{xy} = \iiint\limits_Dz\delta dV$$

**Centro de Massa**
$$    \bar{x}=\frac{M_{yz}}{M}  \text{, }\ \  \bar{y}=\frac{M_{xz}}{M} \text{, }\ \ \bar{z}=\frac{M_{xy}}{M}   $$

**Placa Bidimensional**

**Massa:** $ M = \iint\limits_R\delta dA \ \ \text{onde } \delta=\delta(x,y) \text{ é a densidade em }(x,y)$

**Primeiros momentos**

$$M_{y} = \iint\limits_Rx\delta dA \text{, }\ \ M_{x} = \iint\limits_Ry\delta dA$$

**Centro de Massa**
$$    \bar{x}=\frac{M_{y}}{M}  \text{, }\ \  \bar{y}=\frac{M_{x}}{M} \text{, }  $$
